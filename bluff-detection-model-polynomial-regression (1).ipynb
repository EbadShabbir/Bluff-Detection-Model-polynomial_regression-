```python
#importing the libraries
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#importing datasets  
data_set= pd.read_csv('/kaggle/input/polynomial-regressionpositon-salaries-csv/Position_Salaries.csv')  
  
#Extracting Independent and dependent Variable  
x= data_set.iloc[:, 1:2].values  
y= data_set.iloc[:, 2].values  

# Linear regression model
lin_regs = LinearRegression()
lin_regs.fit(x, y)

# Polynomial regression for degree 2
poly_regs_2 = PolynomialFeatures(degree=2)
x_poly_2 = poly_regs_2.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly_2, y)

# Polynomial regression for degree 3
poly_regs_3 = PolynomialFeatures(degree=3)
x_poly_3 = poly_regs_3.fit_transform(x)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(x_poly_3, y)

# Polynomial regression for degree 4
poly_regs_4 = PolynomialFeatures(degree=4)
x_poly_4 = poly_regs_4.fit_transform(x)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(x_poly_4, y)

# Visualising the Linear Regression result
plt.scatter(x, y, color="blue")
plt.plot(x, lin_regs.predict(x), color="red")
plt.title("Bluff Detection Model (Linear Regression)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()

# Visualising the Polynomial Regression result for degree 2
plt.scatter(x, y, color="blue")
plt.plot(x, lin_reg_2.predict(poly_regs_2.fit_transform(x)), color="red")
plt.title("Bluff Detection Model (Polynomial Regression - Degree 2)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()

# Visualising the Polynomial Regression result for degree 3
plt.scatter(x, y, color="blue")
plt.plot(x, lin_reg_3.predict(poly_regs_3.fit_transform(x)), color="red")
plt.title("Bluff Detection Model (Polynomial Regression - Degree 3)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()

# Visualising the Polynomial Regression result for degree 4
plt.scatter(x, y, color="blue")
plt.plot(x, lin_reg_4.predict(poly_regs_4.fit_transform(x)), color="red")
plt.title("Bluff Detection Model (Polynomial Regression - Degree 4)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()

# LINEAR REGRESSION RESULT
lin_pred = lin_regs.predict([[6.5]])  
print(f"Linear Regression Prediction for 6.5: {lin_pred}")

# POLYNOMIAL REGRESSION RESULT (using degree 2)
poly_pred_2 = lin_reg_2.predict(poly_regs_2.fit_transform([[6.5]]))  
print(f"Polynomial Regression (Degree 2) Prediction for 6.5: {poly_pred_2}")

# POLYNOMIAL REGRESSION RESULT (using degree 3)
poly_pred_3 = lin_reg_3.predict(poly_regs_3.fit_transform([[6.5]]))  
print(f"Polynomial Regression (Degree 3) Prediction for 6.5: {poly_pred_3}")

# POLYNOMIAL REGRESSION RESULT (using degree 4)
poly_pred_4 = lin_reg_4.predict(poly_regs_4.fit_transform([[6.5]]))  
print(f"Polynomial Regression (Degree 4) Prediction for 6.5: {poly_pred_4}")
```
